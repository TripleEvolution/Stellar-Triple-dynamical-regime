# Welcome!
Have you ever come across a stellar triple and wondered: is this system stable?  
 Are three-body dynamics important for this system? This webinterface can help you out.

Using this notebook you can find out in which dynamical regime the triple resides. After filling in the parameters of your triple you can create a figure similar to Fig.4a in Toonen et al. 2019. 
The black dotted lines represent constant Lidov-Kozai timescales of [10^4,10^5,10^6, 10^7] times the period of the outer orbit. The higher-order (octupole) term of the three-body interaction is important below the green dash-dotted line. The triple is in the semisecular regime below the blue dashed line. The system is not dynamically stable below the red solid line. For more information on the dynamical processes, see section 2 of Toonen et al. 2019.<br>

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from figure_functions import plot_figure
```

# Fill in the parameters of your triple:<br>
m1: Mass of the most massive star in the inner binary (solar units) <br>
m2: Mass of the least massive star in the inner binary (solar units)<br>
m3: Mass of the outer companion (solar units)<br>


```python
m1 = 1 
m2 = 0.5 
m3 = 0.75
```

incl: Mutual inclination between the inner and outer orbit (rad). This is required to assess the dynamical stability of the system. Retrograde orbits are more stable than prograde orbits.<br>

```python
incl = np.pi
```

# Overplotting your triple 
To overplot your triple in the figure, set bool to True



```python
OVERPLOT = False
```

Orbital separation of the inner binary


```python
a_in = 10
```

Orbital separation of the outer binary (units same as a_in)


```python
a_out = 1000
```

If needed, convert orbital periods (in seconds) to separations:

```python
P_in = a_in**1.5 * np.sqrt((m1+m2)/(m1+m2+m3))
P_out = a_out**1.5 * np.sqrt((m1+m2)/(m1+m2+m3))
```

Eccentricity of the outer orbit and the uncertainty on it


```python
e_out = 0.2
e_out_error = 0.05
label = 'Algol'

System1 = [a_in, a_out, e_out, e_out_error,label]
```

You can use a similar structure for overplotting multiple objects.<br>
Do note that the boundaries of the dynamical regimes change with mass.<br>

```python
System2 = [10,900,0.4,0.06,'Test']
plot_objects = [System1,System2]
```

# Here you can play around with the boundaries of the different dynamical regimes:<br>

The importance of the octupole term in the secular approximation increases with epsilon_oct (Eq.4)<br>
Generally it is assumed that the octupole term plays a role in the dynamics of the system when epsilon_oct > 0.001-0.01<br>
Here you can set the critical epsilon_oct used to make the figure.<br>

```python
epsilon_oct_crit = 0.005
```

The boundary of the semi-secular regime depends on the maximum eccentricity that can be reached in the inner binary<br>
If the initial inclination is known, you may use the simplistic approximation:<br>


```python
#e_in_max = np.sqrt(1-5./3.*np.cos(incl)**2)
```

For a lower limit estimate:


```python
# e_in_max = e_in
e_in_max = 0.99
```

# Final Plot

In the following section set 'period=True' as an argument to the plot_figure function for y axis as $\frac{P_{out}}{P_{in}}$ instead of $\frac{a_{out}}{a_{in}}$

```python
#plot_figures(m1,m2,m3,e_in_max,incl,oct_lim,list_of_objects,period=True)
if OVERPLOT == True:
    plot_figure(m1,m2,m3,e_in_max,incl,epsilon_oct_crit,plot_objects,period=True)
else:
    plot_figure(m1,m2,m3,e_in_max,incl,epsilon_oct_crit,[],period=True)
```
